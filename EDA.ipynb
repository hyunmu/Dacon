```
#일단 2019 1월 부터 4월까지 2020년 1월-3월까지 나눈 다음에 비교
# 코로나는 작년 12월 중 중국에서 터지고 한국에서 1월달에 최초환자 발생

#일단 날짜, 시도, 업종, 연령, 금액만 남긴다 여기선

df = data.drop(['HOM_SIDO_NM', 'CARD_CCG_NM' 'SEX_CTGO_CD', 'FLC', 'CSTMR_CNT', 'CNT'], axis =1)
df.head()
```

```
# 각 월로 나누기
df_201901 = df.loc[df['REG_YYMM'] == 201901]
df_201902 = df.loc[df['REG_YYMM'] == 201902]
df_201903 = df.loc[df['REG_YYMM'] == 201903]
df_201904 = df.loc[df['REG_YYMM'] == 201904]
df_201905 = df.loc[df['REG_YYMM'] == 201905]
df_201906 = df.loc[df['REG_YYMM'] == 201906]
df_201907 = df.loc[df['REG_YYMM'] == 201907]
df_201908 = df.loc[df['REG_YYMM'] == 201908]
df_201909 = df.loc[df['REG_YYMM'] == 201909]
df_201910 = df.loc[df['REG_YYMM'] == 201910]
df_201911 = df.loc[df['REG_YYMM'] == 201911]
df_201912 = df.loc[df['REG_YYMM'] == 201912]
df_202001 = df.loc[df['REG_YYMM'] == 202001]
df_202002 = df.loc[df['REG_YYMM'] == 202002]
df_202003 = df.loc[df['REG_YYMM'] == 202003]
```

```
#합치기, 2019 01- 04, 2020 01-03 끼리
frames_1 = [df_201901, df_201902, df_201903, df_201904]
frames_2 = [df_202001, df_202002, df_202003]

df_2019 = pd.concat(frames_1)
df_2020 = pd.concat(frames_2)
```

```
# 금액 합치기
df_2019 = df_2019.groupby(['REG_YYMM', 'CARD_SIDO_NM', 'STD_CLSS_NM']).sum().reset_index(drop=False)
df_2020 = df_2020.groupby(['REG_YYMM', 'CARD_SIDO_NM', 'STD_CLSS_NM']).sum().reset_index(drop=False)
```

```
#2019년도 데이터
submission = pd.read_csv('drive/My Drive/JEJU/submission.csv', index_col=0)
submission = submission.loc[submission['REG_YYMM']==202004]
submission = submission[['CARD_SIDO_NM', 'STD_CLSS_NM']]
REG_YYMM9 = np.sort(df_2019['REG_YYMM'].unique())
REG_YYMM9

AMTS = []
for REG_YYMM in REG_YYMM9:
    df1 = df_2019.loc[df_2019['REG_YYMM']==REG_YYMM]
    df1 = df1[['CARD_SIDO_NM', 'STD_CLSS_NM', 'AMT']]
    tmp = submission.merge(df1, left_on=['CARD_SIDO_NM', 'STD_CLSS_NM'], right_on=['CARD_SIDO_NM', 'STD_CLSS_NM'], how='left')
    tmp = tmp.fillna(0)
    AMT = list(tmp['AMT'])
    AMTS.append(AMT)
AMTS = np.array(AMTS)



def plot_graph(dt, AMTS):
    for i in range(dt, len(AMTS)):
        tmp1 = np.log1p(AMTS[i-dt])
        tmp2 = np.log1p(AMTS[i])
        corr = np.corrcoef(tmp1, tmp2)[0][1]
        plt.title('Correlation: %.4f' % (corr))
        plt.scatter(tmp1, tmp2, color = 'k', alpha =0.1)
        plt.xlabel('Log AMT, %s' % (REG_YYMM9[i - dt]))
        plt.ylabel('Log AMT, %s' % (REG_YYMM9[i]))
        plt.xlim(tmp1.min(), tmp1.max())
        plt.ylim(tmp1.min(), tmp1.max())
        plt.show()
```

```
#1개월 차이
plot_graph(1, AMTS)
```

```
# 2개월 차이
plot_graph(2, AMTS)
```

```
# 3개월 차이

plot_graph(3, AMTS)
```
